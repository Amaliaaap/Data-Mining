**1. Persiapan Data**


```python
# Mengimport library yang digunakan (Pandas)
import pandas as pd
```

```python
# membaca file CSV dan menyimpan datanya ke dalam sebuah DataFrame.
df = pd.read_csv('/content/drive/MyDrive/Data Mining/movie_sample_dataset.csv')
df.head()
```

**2. Memeriksa Data**



```python
# Menampilkan baris pertama dari dataset
df.head(1)
```

```python
# Memeriksa informasi umum tentang dataset
df.info()
```

```python
import numpy as np

# Memeriksa jumlah missing value di setiap kolom
print(df.isnull().sum())
```

**3. Membersihkan Data**



```python
df.shape
```

```python
df.loc[:, 'color'] = df['color'].str.strip().str.capitalize()
df.loc[:, 'country'] = df['country'].str.strip().str.upper()
df = df[(df['duration']>=0)&
              (df['gross']>=0)&
              (df['title_year']>=0)&
              (df['budget']>=0)&
              (df['imdb_score']>=0)&
              (df['movie_facebook_likes']>=0)]
df = df.replace('N/A', pd.NA)
df.dropna(inplace=True)
df
```

```python
# Mengecek apakah masih ada Missing data
print(df.isnull().sum())
```

4. Transformasi Data

```python
# Mengecek apa tipe data awal
df.dtypes
```

```python
# Mengubah tipe data
df['gross'] = df['gross'].astype('int')
df['budget'] = df['budget'].astype('int')
df.dtypes
```

```python
df['genres'].str.split()
df['genres'].str.split(pat='|', expand = True)
```

```python
df['actors'].str.split()
df['actors'].str.split(pat=',', expand = True)
```

5. Penyimpanan Data

```python
df.to_csv('movie_dataset_cleaned.csv', index=False)
```
